## 10. Handling Missing Values

Treatment strategy:
- Numerical columns: fill missing values using **median**
- Categorical columns: fill missing values using **mode**

### Why median for numerical columns?

Median is preferred because agricultural data can contain outliers.  
Mean can be affected by extreme values, but median is more robust.

### Why mode for categorical columns?

Mode represents the most frequent category, so it is commonly used for categorical missing value treatment.

```python
for col in num_cols: #d
    df[col] = df[col].fillna(df[col].median())

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

df.isnull().sum()
```

After treatment, missing values are checked again.  
If all values are zero, it means missing values have been successfully handled.

## 14. Outlier Treatment Using IQR Capping

We use the IQR method to cap extreme values.

### Why capping instead of removing?

Removing outliers may reduce dataset size and remove important records.  
Capping controls extreme values while keeping the records.

# Detecting outliers uisng IQR Method

```python
Q1 = df['yield'].quantile(0.25)
Q3 = df['yield'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['yield'] < lower) | (df['yield'] > upper)]

print('Total Outliers in yield column :', outliers.shape[0])
```

```python
def cap_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    outlier_count = ((data[column] < lower_limit) | (data[column] > upper_limit)).sum()

    data[column] = np.where(data[column] < lower_limit, lower_limit,
                   np.where(data[column] > upper_limit, upper_limit, data[column]))

    return outlier_count, lower_limit, upper_limit

outlier_report = []

for col in num_cols:
    if col != "crop_year":
        count, lower, upper = cap_outliers_iqr(df, col)
        outlier_report.append([col, count, lower, upper])

outlier_report_df = pd.DataFrame(outlier_report, columns=["column", "outlier_count", "lower_limit", "upper_limit"])
outlier_report_df
```

### Observation

Outliers were detected and capped using the IQR method.

### Explanation

Values below the lower limit were replaced with the lower limit.  
Values above the upper limit were replaced with the upper limit.  
This reduces the effect of extreme values while preserving the data.
